## Parte 2.1: Extracción del contenido de la página (demostración de código)

````python
# Primero importaremos la libreria "requests" que nos ayudará a conectar con la API de Wikipedia
import requests

#Check verde al costado indica que el código se ejecutó sin problemas
````

La lógica es la siguiente:

1.   Usar la función "get" para conectar con la API de Wikipedia. Especificando el enlace (URL) y los diferentes parámetros requeridos.
2.   Almacenar la respuesta del servidor en un objeto llamado "respuesta".
3. Visualizar la respuesta utilizando el comando .json




````python
# Primero crearemos el objeto "respuesta" que almacenará la respuesta que nos de el servidor (AKA la información de la página)
respuesta = requests.get(url = "https://es.wikipedia.org/w/api.php", #este es el endpoint para Wikipedia en español, es decir desde donde extraera la info
                         params= {
                              "action": "parse", #Acción que le indica al API qué acción ejecutar. En este caso "analizar"
                              "page": "Época del terrorismo en el Perú", # El nombre de la página de Wikipedia de donde queremos extraer la data
                              "format": "json", #El formato de la respuesta. En este caso es JSON, podría también ser XML u otros
                              "prop": "text", #indica que nos de el contenido en texto versión HTML de la página
                              "redirects": 1 #para evitar que redireccione la pagina
                         })

respuesta #el output nos muestra "Response [200]" que significa que la respuesta fue exitosa.
# Sin embargo, no podemos ver el contenido del objeto. Para ello, debemos hacer un segundo paso, para leer la respuesta que esta en formato json
````

````python
# La forma más rápida es imprimir el "texto" del objeto respuesta
print(respuesta.text)
````

````python
# Una segunda forma es con la función ".json" que permite guardarlo en un nuevo objeto (llamado data) y luego imprimirlo.
# La ventaja de trabajar con formato JSON es que es más amigable para el ojo humano y un formato más universal (máquinas).
# Adicionalmente, con esta segunda forma, hay más elementos que aparecen, por ejemplo las imágenes.
data = respuesta.json()  # La repuesta
print(data)
````

La solicitud que realizamos a la API nos permitió acceder a la información requerida. Sin embargo, si tuviesemos que trabajar extrayendo una mayor cantidad de data de diferentes páginas, sería importante estandarizar este proceso y automatizarlo. Para ello, definiremos una función que nos permita extraer el contenido de la API solo con una línea de código especificando únicamente el nombre de la página en Wikipedia.

````python
# Definimos una función llamada 'extraer_contenido' que tenga como único argumento el titulo de la pagina web que requerimos.
def extraer_contenido(titulo):
  # Paso 1: Definimos el URL y los parámetros que queremos utilizar
  url = "https://es.wikipedia.org/w/api.php"
  parametros = {
      "action": "parse",
      "page": titulo,
      "format":"json",
      "prop": "text"
  }

  # Paso 2: La función enviará la solicitud a la API dado los parámetros especificados en el Paso 1
  respuesta = requests.get(url, params=parametros)

  # Paso 3: Leer la respuesta de la API que se encuentra en formato JSON en un dictionario estructurado en Python
  data = respuesta.json()

  # Paso 4: Extraer el código html del formato JSON a un objeto llamado pagina_html y luego imprimirlo
  page_html = data.get("parse", {}).get("text", {}).get("*", "")
  return page_html
````

````python
# Ahora que hemos creado nuestra función, podemos aplicarla en solo una linea de código!
extraer_contenido("Época del terrorismo en el Perú")

#Probemoslo en un visualizador de HTML: https://html.onlineviewer.net/
#Copien y peguen el output que figura aquí en esa página y podemos comprobar que se extrajo.
````

````python
# Finalmente, ahora que accedimos a la información hay que guardarla.
# Para ello crearemos una función(llamada html_archivo) que guarde el output especificando dos argumentos.
# 1) será el objeto (en Python) con el contenido, y 2) el segundo el nombre del archivo .

def html_archivo(contenido_html, nombre_archivo="page.html"):
    with open(nombre_archivo, "w", encoding="utf-8") as file: #especificamos que guarde con encoding utf-8 que lee caracteres especiales
        file.write(contenido_html)
    print(f"HTML content saved to {nombre_archivo}") #le pedimos a la función que cuando guarde el objeto imprima un mensaje que indique que se guardó exitosamente

# Antes de aplicar la función primero hay que guardar el contenido en un objeto, al que llamaremos CAI_Peru
CAI_Peru = extraer_contenido("Época del terrorismo en el Perú")

# Utilizamos la función que creamos para guardar el objeto en un archivo html
html_archivo(CAI_Peru, "CAI_Peru_Wikipedia.html")

# Ahora podemos comprobar que el archivo se guardo correctamente. Descarguemoslo porque eso lo utilizaremos en la siguiente sesión!
````

## Parte 2.2 - Ejercicio 1: Extraer historial de revisiones de usuarios



Ahora la idea es que ustedes puedan familiarizarse más con la extracción de información usando la API extrayendo otro tipo de información que la API brinda. <br></br> Wikipedia registra el historial de revisiones que los/as usuarios ejecutan en cada una de las páginas. A cada una de estas modificaciones se les conoce como revisiones y se puede acceder a su historial y visualizar como esta ha cambiado a través del tiempo. <br></br> Cada revisión tiene un identificador único (ID), el nombre del usuario que realizó la modificación, el tipo de modificación que se hizo, entre otros datos relevantes que podríamos utilizar para ver su evolución en el tiempo. <br></br>  Para este ejercicio, se les pide que con la ayuda de la [documentación oficial de la API de Wikipedia](https://www.mediawiki.org/wiki/API:Allrevisions) y herramientas externas (StackOverflow, ChatGpt, etc.) puedan extraer el historial de revisiones para esta página. <br></br> Para entender mejor el historial de revisiones en Wikipedia pueden revisar [siguiente página](https://en.wikipedia.org/wiki/Help:Page_history#) de Wikipedia, lo coloco en la versión en inglés ya que contiene información más detallada que su versión en español


````python
# En estas líneas de código podrán encontrar la respuesta al ejercicio
# Definimos una función que extraiga el historial de la página de Wikipedia requerida
def extraer_historial_revision(titulo):
 url = "https://es.wikipedia.org/w/api.php"
 parametros = {
       "action": "query", # la acción ya no es analizar, sino "buscar"
       "format": "json",
       "prop": "revisions", # Este parámetro indica que se busquen las revisiones a la página
       "titles": titulo,
       "rvprop": "timestamp|user|comment|ids|size", #las propiedades de cada revisión que pedimos extraer
       "rvlimit": "max", # parámetro que indica que extraiga el número máximo de revisiones por cada solicitud que se haga a la API
       "rvdir": "newer" # el orden de las revisiones será primero la más antigua
 }

#Paso 2: Conectar con la API
 respuesta = requests.get(url, params=parametros)
 data = respuesta.json()

 return data

# Aplicar la función
extraer_historial_revision("Época del terrorismo en el Perú")
````

````python
# Segunda parte de la respuesta
# Este output puede llegar a ser un poco desordenado (a nuestra vista humana),
# pero con algunas modificaciones en el codigo se puede convertir en una base de datos tal como la conocemos (matriz)
# Para ello crearemos otra función que además de extraer el historial también convierta el output en una base de datos (matriz).
def extraer_historial_revision_df(titulo):
 url = "https://es.wikipedia.org/w/api.php"
 parametros = {
       "action": "query",
       "format": "json",
       "prop": "revisions",
       "titles": titulo,
       "rvprop": "timestamp|user|comment|ids|size",
       "rvlimit": "max",
       "rvdir": "newer"
 }

 # Paso 2: Conectar con la API
 respuesta = requests.get(url, params=parametros)
 data = respuesta.json()


 # Paso 3: Extraer las listas en format json
 paginas = data.get('query', {}).get('pages', {}) # creamos un diccionario llamado paginas que busca dentro del campo "query" y "pages" del output en formato JSON
 revisiones_por_pagina = [] #creamos una lista vacia para almacenar la información que extraeremos de la respuesta en JSON obtenida

 # Paso 4: Crear un loop que extraiga todas las revisiones por página
 # Cada página posee un identificador único (ID) y tiene la info requerida (información de las revisiones)
 # lo que hace el loop es primero buscar si es que existe el parametro "revisions" (si está presente, significa que hay revisión) en la informacion de la pagina:
 # y si es que está presente, entonces extrae la información ("revisions") y las almacena en la lista vacia "revisiones_por_pagina"
 for id_pagina, info_pagina in paginas.items():
   if "revisions" in info_pagina: # condicional que verifica si es que la pagina tiene revisiones, si cumple con ello, aplica el siguiente código
     revisiones_por_pagina = info_pagina["revisions"] #si la pagina tiene revisiones, añade esas revisiones a la lista vacia "revisiones_por_pagina"


# Paso 5: Crear una lista completa de revisiones
 lista_completa_revisiones = [] # Lista vacia para almacenar las revisiones


# Paso 6: Loop que vaya por cada revision y extraiga cada una de las siguientes variables (revid, timestamp, etc.)
# y lo añada a un nuevo objeto llamado "data_revisiones" que tendra cada revision con sus respectivas variables
 for cada_revision in revisiones_por_pagina:
   data_revisiones = {
       "revid": cada_revision.get("revid"),
       "parentid": cada_revision.get("parentid"),
       "timestamp": cada_revision.get("timestamp"),
       "user": cada_revision.get("user"),
       "size": cada_revision.get("size"),
       "comment": cada_revision.get("comment")
   }
   lista_completa_revisiones.append(data_revisiones) #añadir estas variables a la lista vacia que creamos anteriormente

# Paso 6: Convertir la lista de revisiones en una base de datos (matriz). Para ello usaremos la libreria pandas
 base_datos = pd.DataFrame(lista_completa_revisiones)


 return base_datos

# Debemos importar la libreria pandas para la ultima parte de la funcion
import pandas as pd


# Aplicar la función creada
extraer_historial_revision_df("Época del terrorismo en el Perú")


#El resultado son 500 revisiones distintas...
````

## 3.5. Extraer varios titulares (loop)

Con lo aprendido previamente, el reto es  extraer todos los titulares de las últimas 3 páginas, de modo que podamos hacer una base de datos más grande de la cual extrar información y responder a la pregunta que tenemos.

Para ello, primero debemos definir una palabra de búsqueda que nos permita acceder a las noticias que refieran a "migrantes venezolanos". Para ello utilizaremos la palabra "venezolano".

Si colocamos esta palabra en el motor de búsqueda de la página web del Correo, entonces podemos entender la lógica detrás del motor de búsqueda y modificarlo para poder generar una base de datos que nos permita extraer las noticias de las últimas 5 páginas.

> https://diariocorreo.pe/buscar/venezolano/todas/descendiente/1/

Cada página muestra 30 noticias, por lo que si extraemos las primeras 5 páginas, tendremos un aproximado de 150 titulares. Para ello, utilizaremos la siguiente lógica:

1. Crear una base de datos con los 5 enlaces (loop)
2. Identificar el titular en el árbol HTML
3. Diseñar una función de extracción
5. Guardar el contenido en una base de datos (matriz)



````python
# Paso 1: Crear una base de datos con los enlaces para las primeras 5 noticias

enlace_base= 'https://diariocorreo.pe/buscar/venezolano/todas/descendiente/' #definimos en enlace base

pagina_inicial = 1
pagina_final = 5

# Creamos una lista vacia para almacenar todos estos enlaces
enlaces_correo = []

# Loop para agregar el numero de cada página
for numero_pagina in range(pagina_inicial, pagina_final + 1): # por cada pagina en el rango de 1 al 5 haz lo siguiente
  url = f'{enlace_base}{numero_pagina}/' #junta el 'enlace_base' con 'el numero de la pagina actual' y añadele "/". guardalo en un objeto llamado "url"
  enlaces_correo.append(url) # Añade el url anterior a la lista vacia 'enlaces_correo'

# Mostramos nuestra lista
print(enlaces_correo)
````

**Paso 2: Identificar el titular en el código HTML**

Ahora que ya tenemos nuestra base de datos con las 5 primeras páginas de búsqueda de noticias, es momento de identificar el tag que contiene el titular de cada noticia:

> < a itemprop="url" class="story-item__title block overflow-hidden primary-font line-h-xs mt-10" href="/edicion/huancayo/huancayo-ocultan-mas-de-2-toneladas-de-acido-sulfurico-en-ladrillos-y-cemento-noticia/">Huancayo: Ocultan en camión más de 2 toneladas de ácido sulfúrico en ladrillos y cemento</a >

Con este código, identificamos que para acceder al titular, debemos acceder por medio del atributo "class" con el siguiente valor: "class="story-item__title block overflow-hidden primary-font line-h-xs mt-10"

````python
# Paso 3: Definir una función de extracción

# Definir una base de datos vacias para almacenar los titutares
correo_df = pd.DataFrame(columns=['titular'])

# Diseñar una función que se diriga a cada url y extraiga los titulares y luego los añada dentro de una base de datos
def extraer_titular(lista_url):
  global correo_df
  for url in lista_url:
    respuesta_servidor = requests.get(url) # conectar con el servidor
    pagina_html = BeautifulSoup(respuesta_servidor.text, 'html.parser') # transformarlo en objeto BeautifulSoup
    titulares = pagina_html.find_all('a', class_ = 'story-item__title block overflow-hidden primary-font line-h-xs mt-10') #Extrer el tiutlar

    #Añadir en una base de datos
    for titular in titulares:
      correo_df = pd.concat([correo_df, pd.DataFrame({'titular': [titular.get_text()]})], ignore_index=True)

    # En caso sea una base mas grande se puede añadir una pausa entre 1 y 3 segundos para no dañar el servidor
    #time.sleep(random.uniform(1, 3)) # se necesita libreria 'time' y 'random'
````
